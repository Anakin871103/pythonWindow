```python
import requests
url = "https://www.vscinemas.com.tw/vsweb/film/index.aspx"
response = requests.get(url)
if response.status_code == 200:
    print("下載成功")
    print(response.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text,'html.parser')
movies=soup.find('ul',class_='movieList')
```

```python
for liTag in movies.find_all('li'):
    print(liTag.find('section',class_='infoArea').h2.a.string)
```
