```python
import os
listdir = os.listdir('./assets')
sorted_list_dir= sorted(listdir)
file_name = sorted_list_dir[-1]
abs_file_name = os.path.abspath('./assets/'+file_name)
abs_file_name
```

```python
import csv

with open(abs_file_name,'r',encoding='utf-8') as file:
    rows = csv.reader(file)
    datalist = list(rows)

for item in datalist:
    print(item)
```
